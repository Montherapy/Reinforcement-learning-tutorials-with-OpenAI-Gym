# A2C (Advantage Actor-Critic)

## Introduction of A2C
The REINFORCE algorithm uses total return for the policy gradient. This has the disadvantage of having to wait until the end of the episode and the variance of the total return term is large. If the equation is rearranged, the policy gradient can use the Q function instead of the total return. Subtracting the baseline (the behavior-independent term) from the Q term yields a mathematically equivalent policy gradient with lower variance. In A2C, the advantage term is used by subtracting the state value function.

## Points of traning

#### Reward normalization 
Since the reward is used to calculate the gradient, it is better to use an appropriate scale value in order to train parameters well. Normalizing the reward helps to control the magnitude of the gradient. In general, variance can also be adjusted by standardising the reward. In this example, after finding out the approximate distribution of rewards by recording the rewards output from the training process, the real training is performed using normalized reward values. Fortunately, this example is a continuing problem, so even the overall mean shift of reward does not affect the optimal policy. If it was an episodic problem, only positive scaling is possible. The recorded reward set was saved in advance.
#### Continuing problem
In the Pendulum-v0 environment, the episode does not end unless time runs out. Therefore, unlike the Cartpole environment, it is reasonable not to use variable "done" when updating the value function.
#### Setting the range of standard deviation of policy
When calculating the log-Gaussian policy for calculating the policy gradient, the variance of the policy distribution is included in the denominator. In this case, if the value of the variance is too small, the loss may diverge and the training may fail. To prevent this, we forcefully set the range of standard deviation, which is the output of the neural network.

## Pendulum environment

To use Actor Critic as a probabilistic policy neural network method, we chose the Pendulum tutorial which has continuous action space. <br>
According to [1], the state, action, reward is as follows:
- State : cosθ, sinθ, α 
- action : τ 
- reward : -θ^2 - 0.1α^2 - 0.001τ
<br>
where θ is an angle between the upward direction and the bar, α is an angular velocity, and τ is an torque of the rotational joint. 

We can check the observation and action spaces.

```python
import gym
import torch
import torch.nn as nn
from torch.nn import functional as F
import matplotlib.pyplot as plt
import numpy as np
from collections import namedtuple

env = gym.envs.make("Pendulum-v0") 
print("observation space range (Cos(theta), Sin(theat), Angular velocity): {} ~ {} ".format(env.observation_space.low,env.observation_space.high))
print("action space range (Joint torque) : {} ~ {} ".format(env.action_space.low, env.action_space.high))
```

## Actor critic network

```python
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
GAMMA = 0.95
BATCH = 16
ACTOR_LR = 0.0001
CRITIC_LR = 0.001
STD_BOUND = [1e-3, 1] # To make the log_prob of pi(a|s) not diverge.
```

```python
class Actor(nn.Module):
    def __init__(self):
        super(Actor, self).__init__()
        self.fc1 = nn.Linear(obs_size, 32)
        self.fc2 = nn.Linear(32, 32)
        self.fc3 = nn.Linear(32, 32)
        self.mu = nn.Linear(32, act_size)
        self.log_std = nn.Linear(32, act_size)

    def forward(self, obs):
        x = F.relu(self.fc1(obs))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        mu = torch.tanh(self.mu(x)) * 2  
        std = F.softplus(self.log_std(x))
        std = torch.clip(std, STD_BOUND[0], STD_BOUND[1])
        act = torch.normal(mu, std)
        
        return mu, std, act
    
    def get_loss(self, obs, act, adv):
        mu, std, _ = self.forward(obs)
        pi = torch.distributions.Normal(mu, std)
        log_prob = pi.log_prob(act)
        actor_loss = - log_prob * adv.detach()
        
        return actor_loss.mean()
    
class Critic(nn.Module):
    def __init__(self):
        super(Critic, self).__init__()
        self.fc1 = nn.Linear(obs_size, 32)
        self.fc2 = nn.Linear(32, 32)
        self.fc3 = nn.Linear(32, 32)
        self.fc4 = nn.Linear(32, 1)

    def forward(self, obs):
        x = F.relu(self.fc1(obs))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        v = self.fc4(x)

        return v
    
class A2C:
    def __init__(self, actor, critic):
        self.actor = actor
        self.critic = critic
        self.transition = namedtuple("Transition" , ['s', 'a', 'r', 's_'])
        self.buffer = []
        self.criterion = nn.MSELoss()
        self.actor_optimizer = torch.optim.Adam(self.actor.parameters(), ACTOR_LR)
        self.critic_optimizer = torch.optim.Adam(self.critic.parameters(), CRITIC_LR)

    def get_action(self, s, sampling=True):
        s_tensor = (torch.from_numpy(s)).float().to(DEVICE) 
        mu, std, act = self.actor(s_tensor)
        if sampling:
            action = act.item()
        else :
            action = mu.item()
        action = np.clip(action, -2., 2.)
        return action
    
    def stack_buffer(self, sample):
        self.buffer.append(self.transition(*sample))    
        
    def update(self, normalizing_reward=True):
        s = torch.tensor([e.s for e in self.buffer]).reshape(BATCH, obs_size).float().to(DEVICE)
        a = torch.tensor([e.a for e in self.buffer]).reshape(BATCH, act_size).float().to(DEVICE) 
        r = torch.tensor([e.r for e in self.buffer]).reshape(BATCH, 1).float().to(DEVICE)
        s_ = torch.tensor([e.s_ for e in self.buffer]).reshape(BATCH, obs_size).float().to(DEVICE)
        self.buffer.clear()
        
        # TD_error 
        if normalizing_reward : 
            r = (r-r_mean)/r_std
#         target = (r + (1-d) * GAMMA * self.critic(s_)).detach()
        target = (r +  GAMMA * self.critic(s_)).detach()
        v = self.critic(s)
        critic_loss = self.criterion(v, target)
        advantage = target - v
        actor_loss = self.actor.get_loss(s, a, advantage)
        
        self.actor_optimizer.zero_grad()
        self.critic_optimizer.zero_grad()
        actor_loss.backward()
        critic_loss.backward()
        self.actor_optimizer.step()
        self.critic_optimizer.step()

        return actor_loss.detach().cpu().numpy()
```

## Visualization 

```python
 class Graph():
    def __init__(self):
        self.max_num_data = 200
        
        self.fig, self.ax = plt.subplots(1, 2, figsize=(12, 4))
        self.loss_graph, = self.ax[0].plot(1)
        self.reward_graph, = self.ax[1].plot(1)

        self.ax[0].set_title("Loss")
        self.ax[0].set_xlim(0,1)
        self.ax[0].set_ylim(0,2)
        self.ax[0].set_xlabel('step')
        self.ax[0].set_ylabel('loss')

        self.ax[1].set_title("Cumulative reward")
        self.ax[1].set_xlim(0,10)
        self.ax[1].set_ylim(0,2)
        self.ax[1].set_xlabel('episode')
        self.ax[1].set_ylabel('reward')

    
    def update_loss(self, loss_list):
        while len(loss_list) > self.max_num_data:
            loss_list = [data for idx, data in enumerate(loss_list) if idx %2 == 0]
        
        x = [i[0] for i in loss_list]
        y = [i[1] for i in loss_list]
        
        self.ax[0].set_xlim(0,x[-1])
        self.ax[0].set_ylim(np.min(y)*0.9,np.max(y)*1.1)

        self.loss_graph.set_xdata(x)
        self.loss_graph.set_ydata(y)        

        self.fig.canvas.draw()
        self.fig.canvas.flush_events()     
        
    def update_reward(self, reward_list):
        while len(reward_list) > self.max_num_data:
            reward_list = [data for idx, data in enumerate(reward_list) if idx %2 == 0]
        
        x = [i[0] for i in reward_list]
        y = [i[1] for i in reward_list]
        
        self.ax[1].set_xlim(0,x[-1])
        self.ax[1].set_ylim(np.min(y)-0.1*abs(np.min(y)), np.max(y)+0.1*abs(np.max(y)))
        self.reward_graph.set_xdata(x)
        self.reward_graph.set_ydata(y)        

        self.fig.canvas.draw()
        self.fig.canvas.flush_events()    
```

## Start training

```python
obs_size = env.observation_space.shape[0]
act_size = env.action_space.shape[0]
actor = Actor().to(DEVICE)
critic = Critic().to(DEVICE)
agent = A2C(actor, critic)
```

```python
%matplotlib notebook
plt.ion()
graph = Graph()
loss_list = []
reward_list = []
loss_record_term = 100
max_steps = 500000
num_update = 0
episode = 0
step = 0
while step < max_steps:
    episode += 1
    cum_reward = 0
    state = env.reset()
    
    while True :
        step += 1
#         env.render()
        action = agent.get_action(state)
        next_state, reward, done, info = env.step([action])
        sample = [state, action, reward, next_state]
        agent.stack_buffer(sample)
        cum_reward += reward    

        state = next_state
                
        # update
        if len(agent.buffer) == BATCH:
            loss = agent.update()
            num_update += 1
        # record loss
        if step % loss_record_term == 0: 
            loss_list.append([step,loss])
            graph.update_loss(loss_list)
            #print("num_update, loss , epsilon : " ,num_update, loss, agent.epsilon)

        # terminated
        if done:
            reward_list.append([episode, cum_reward])
            graph.update_reward(reward_list)
            #print("episode: {}, total reward: {}".format(episode, cum_reward))
            break
        step += 1
env.close()
```
